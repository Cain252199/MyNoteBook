## 冯·诺伊曼结构
* 组成框图：
    * <img style="float: center;" src="./冯诺依曼型机器组成框图.jpg" width=400 height=400>
* 特征：
    1. 存储器是字长固定的，顺序线性编制的一维结构
    2. 存储器提供可按地址访问的一级地址空间，每个地址是唯一定义的
    3. **由指令形式的低级机器语言驱动**
    4. 指令的执行是顺序的
    5. **机器以运算器为中心**
* 40年来的改进：
    1. 计算机系统结构从基于串行算法改变为适应并行算法  ---- 向量计算机，并行计算机，多处理机
    2. 高级语言与机器语言的语义距离缩小               ---- 面向高级语言机器和直接执行高级语言机器
    3. 硬件子系统与操作系统和数据库管理系统软件相适应  ---- 面向操作系统机器和数据库计算机
    4. 从传统的指令驱动型改变为数据驱动型和需求驱动型  ---- 数据流机器和归约机
    5. 出现了各种专用计算机
    6. 为了获得高可靠性研制了容错计算机
    7. 出现了分布式计算机
    8. 出现了与大规模、超大规模集成电路相适应的计算机系统结构
    9. 出现了处理非数值化信息的智能计算机

## 软件对系统结构的影响
* **软件必须具有可兼容性，可移植性**
* 解决软件的可兼容性问题：
    1. 采用系列机方法  
        * 采用具有相同的系统结构，但具有不同组成和实现的一系列不同型号的机器
    2. 采用模拟和仿真方法
        * 模拟方法：用软件方法在一台现有的计算机上实现另一台计算机的指令系统
        * 仿真方法：用微程序的直接解释另一种机器指令系统的方法
        * 模拟程序放在主存中，仿真程序放在控存中
    3. 采用同一的高级语言
