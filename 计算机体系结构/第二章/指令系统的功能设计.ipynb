### 指令系统的功能设计
* 引论
    * 硬件指令：速度高，价格贵，灵活性差
    * 软件指令：速度低，价格便宜，灵活性好
    * 设计指令系统时，功能方面的基本要求：
        1. 完整性：具备基本的指令种类（计算机的5类基本指令）
        2. 规整性：
            * 对称性：对所有寄存器同等对待，对数据存储设备的使用和操作码的设置等都要对称
            * 均匀性：对不同的数据类型、字长、操作种类和数据存储设备都要设置相同的指令
        3. 高效率：指令的执行速度要快，使用频率要高；各类指令之间要有一定的比例
        4. 兼容性：在同一系列机内指令系统不变（或适当增加）

### 基本指令系统
* 5类基本指令
    1. 数据传送类指令
        * 决定因素：
            * 数据存储设备的种类：通用寄存器，主存储器，堆栈
            * 数据传送的单位：字，字节，位，数据块等 
            * 采用的寻址方式
            
        * 指令种类
        
        <img src = "./传送指令.png" width = 80%>
    2. 运算类指令
        * 设计时考虑四种因素的组合：
            1. 操作种类：加减乘除，与或非，异或，比较，移位，检索，转换，匹配，清除，置位等
            2. 数据表示：定点数，浮点数，逻辑数，十进制数，字符串，定点向量等
            3. 数据长度：字，半字，双字，字节，位，数据块等
            4. 数据存储设备：通用寄存器，主存储器，堆栈等
        * 特殊操作：
            * 移位：移位方向（L/R），移位种类（算术移位A，逻辑移位L，循环移位R），移位长度（单字S，双字D）
            * 位操作：置位，清位，位测试，找位等
            * 字符串操作：比较，查找，匹配，转换等
    3. 程序控制类指令
        * 转移指令
            * 无条件转移：局部无条件转移（相对寻址），全局无条件转移
            * 有条件转移
                * 转移条件：全零(Z)，正负号(N)，进位(C)，溢出(V)及他们的组合
        * 程序调用和返回指令
            * CALL/RETURN/中断控制指令(开中断，关中断，改变屏蔽，中断返回，自陷等0)
        * 循环控制指令   --- 对循环程序的优化
    4. 输入输出指令
        * 采用单一的直接寻址方式
        * 主要指令有：启动，停止，测试，控制设备，数据输入，数据输出
        * 在多用户或多任务下，输入输出指令属于特权指令
        * 当输入输出设备与主存储器共用一个零地址空间时，可以不设置输入输出指令
    5. 处理机控制和调试指令
        * 处理机状态的切换-----（处理机至少有两个状态）
        * 硬件和软件的调试指令
            * 硬件调试指令：钥匙位置、开关状态的读取寄存器和主存单元的显示等
            * 软件调试指令：断点的设置、跟踪、自陷井指令等

### 复杂指令系统(CISC)
*增强指令的功能，设置一些功能复杂的指令，把一些原来有软件实现的，常用的功能改用硬件的指令系统来实现*
* 目的程序的优化
    * 目的程序：由机器指令直接组成，直接在处理机中执行
    * 优化指标：缩短程序的长度，缩短程序的执行时间
    * 优化方法：
        - 将频率高的指令串用一条新的指令来代替
        - 用硬件加快频率高的指令
    * 优化途径：
        1. 增强数据传送指令的功能
        2. 增强运算型指令的功能
        3. 增强程序控制指令的功能

* 对高级语言和编译程序的支持
    * 目的：缩小高级语言与机器语言的差距，提高整个计算机系统的性能
    * 途径：
        1. 增强对高级语言和编译程序支持的指令的功能
            - 对使用频率高、执行时间长的语句，增强有关指令的功能，或增加专门的指令
        2. 高级语言计算机
            - 把高级语言和机器语言合二为一
