## 计算机系统层次结构
1. 第0级 **硬联逻辑**
    * 计算机的硬件内核，由门，触发器等逻辑电路组成
2. 第1级 **微程序控制**
    * 机器语言：微指令集，由硬件直接执行微指令集编写的微程序  
3. 第2级 **传统机器语言计算机**
    * 机器语言：指令系统，由第1级的微程序对指令系统编写的程序进行解释
4. 第3级 **操作系统**
    * 机器语言：指令系统和操作系统级指令
        * 指令系统的指令由第1级的微程序解释，操作系统级指令由操作系统进行解释。
        * **操作系统**是运行在第2级上的**解释程序**
5. 第4级 **汇编语言计算机**
    * 机器语言：汇编语言，由汇编语言编写的程序先被翻译成第3级或第2级语言，在进行解释
    * 完成翻译的程序叫做汇编程序
6. 第5级 **高级语言计算机**
    * 机器语言：各种高级语言
7. 第6级 **应用程序**

* 第0级机器由硬件实现
* 第1级机器由微程序（固件）实现
* 第2级至第6级机器由软件实现
  * *由软件实现的机器成为虚拟机器*

## 计算机系统结构的定义: computer architecture
由程序设计者看到的一个计算机系统的属性，即概念性结构和功能特性（计算机系统的外特性）

## 计算机的组成与实现
### 计算机的组成任务
   计算机组成的任务是在计算机系统结构确定分配给硬件子系统的功能及其概念结构之后，**研究各个组成部分的内部构造和相互联系**，以**实现机器指令级的各种功能和特性**
   * 相互联系：包括各功能部件的配置、相互连接和相互作用

**计算机系统结构、计算机组成和计算机实现是三个不同的概念。系统结构是计算机系统的软，硬件的界面；计算机组成是计算机系统结构的逻辑实现；计算机实现是计算机组成的物理实现**

## 计算机系统结构的分类
1. 按 **性能与价格** 划分：巨型，大型，中型，小型，微型
2. 按 **用途** 划分：科学计算，事务处理，实时控制，家用
3. 按 **处理机个数和种类** 划分：单处理机，多处理机，并行处理机，关联处理机，超标量处理机，超流水线处理机，SMP（对称多处理机），MPP（大规模并行处理机），机群系统

### 分类方法
    -----《计算机体系结构》第7-8页
1. Flynn分类法 
    * 1966年 M.J.Flynn 提出如下定义：
        * 指令流 -- 机器执行的指令序列
        * 数据流 -- 由指令流调用的数据序列，包括输入数据和中间结果
        * 多倍性 -- 在系统最受限制的元件上同时处于同一执行阶段的指令或数据的最大个数
    * 分类
        * **单指令流单数据流（SISD）**
        * **单指令流多数据流（SIMD）**
        * **多指令流单数据流（MISD）**
        * **多指令流多数据流（MIMD）**
2. 冯式分类法
  * *（冯泽云 1972年提出用**最大并行度**进行分类）*
  * **字串位串（WSBS）**
  * **字并位串（WPBS）**
  * **字串位并（WSBP）**
  * **字并位并（WPBP）**
3. $ H\ddot a ndler$分类法
    * 把计算机的硬件结构分成三个层次，并考虑他们的**可并行-流水处理程度**
    * 三个层次：
        * 程序控制部件（PCU）的个数 k
        * 算术逻辑部件（ALU）或处理部件（PE）的个数 d
        * 每个算术逻辑部件包含基本逻辑线路（ELC）的套数 w；
