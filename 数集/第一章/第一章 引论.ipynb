## 数字集成电路设计中的问题
* **摩尔定律**：在芯片的单位面积上**集成的晶体管数目（集成度）** 大约每1-2年增加一倍
* **等比缩小**：
    * 工艺特征尺寸按每代0.7倍缩小
    * 单片集成的功能按每代2倍增长，芯片成本不显著增加
    * 功能成本按0.5倍下降

## 数字设计的质量评价
* **成本**
    * 固定成本
        1. **与产品销售数量无关的**，如设计花费的时间和人力，芯片制造的用的掩模版等；
        2. 降低设计成本是半导体工业永远面临的主要挑战之一
    * 可变成本
        1. 与产品的产量和芯片的面积成正比
        2. 当 芯片成本：芯片测试成本：封装成本 = 1: 1: 1时，可变成本最低
    * 计算公式
        1.  $  每颗IC的成本 = 每颗IC的可变成本 + \frac{固定成本}{产量}$
        $ \\ $
        2.  $ 可变成本 = \frac{芯片成本+芯片测试成本+封装成本}{最终测试的成品率}$
        $ \\ $
        3.  $ 芯片成本 =  \frac{圆片成本}{每个圆片的新片数 \times 芯片成品率}  = f(芯片面积)^4$
        $ \\ $
        4.  $ 良率 = \frac{每个圆片完好的芯片数}{每个圆片的芯片数} \times 100 \% $
        $ \\ $
        5.  $ 每个圆片的芯片数 = \frac{ \pi \ \  \times \ \  (圆片直径/2)^2}{芯片面积} \ \ \ \ \  - \ \  \frac{\pi \ \ \ \times\ \ \ 圆片直径}{\sqrt{2 \ \ \ \times \ \ \ 芯片面积}}$
        $ \\ $
        6.  $ 芯片的成品率 = (1 + \frac{单位面积的缺陷数 \times 芯片面积}{ \alpha })^{-\alpha} \\
            \alpha 近似等于3，与掩模数量大致成正比$
* **功能性和稳定性**
    0. 噪声来源：**电感耦合，电容耦合，电源地噪声**
    1. **电压传输特性（VTC）**
        * 公式：$V_{out} = f(V_{in})$
        * $ V_{OH} 和 V_{OL}表示额定的高低电压\\
            V_{M}表示门阈值电压或开关电压\\
            V_{IH}表示可接受高电压的最低值\\
            V_{IL}表示可接受低电压的最高值$
            
        * $V_{M} = f(V_M), y = x 与VTC的交点 \\
           V_{IH}和V_{IL}的取值为 dV_{out}/dV_{in} = -1 时，V_{in}的两个取值\\ $
           
        * $V_{IH},V_{IL},V_{OH},V_{OL}的关系图 以及反相器的VTC曲线\\ $
            ![V_{IH},V_{IL},V_{OH},V_{OL}的关系图](./VOH.PNG "V_{IH},V_{IL},V_{OH},V_{OL}的关系图") 
            ![反相器的VTC曲线](./VTC.PNG "反相器的VTC曲线")
         
    2. **噪声容限**
       * $NM_L = V_{IL} - V_{OL} \\
          NM_H = V_{OH} - v_{IH}$
       * $NM_L和NM_H$ 应当大于零，且越大越好
       * 示意图
       
       <img src = "./噪声容限.png" width = 25%>
       
    3. **可再生性**
        * 保证一个受干扰的信号在通过若干个逻辑级后逐渐收敛回到额定电平中的一个
    4. **抗噪声能力**
        * 使只有很小一部分可能由破坏能力的噪声能耦合至重要的电路节点
        * 噪声源的类型：
            * 与信号摆幅$V_{sw}$成正比的噪声 --->$gV_{sw}$
            * 固定噪声 ----->$fV_{Nf}$
        
        * 将总的噪声容限分配给预计的噪声源
    5. **方向性** ---->单方向性，输出对输入的影响很小
    6. **扇入和扇出**
        * 扇出：连接到驱动门输出端的负载门的数量
        * 扇入：该门的输入的数目
        * 负载门的输入电阻必须尽可能的大，驱动门的输出电阻必须尽可能的小
    7. **理想数字门**
        * $ R_i = \infty;R_O = 0; 扇出 = \infty; NM_H = NM_L = \frac{V_{DD}}{2}  $
* **性能**
     * 数字设计的性能常用时钟周期的长短或时钟频率的快慢衡量。
    * 延时：
        * tpHL:输出由高变低翻转的**响应时间**，50%Vin->50%Vout
        * tpLH:输出由低变高翻转的**响应时间**，50%Vin->50%Vout
        * tf:输出由高变低翻转的**转换时间**，90%Vin->10%Vout
        * tr:输出由低变高翻转的**转换时间**，10%Vout->90%Vout
        * 传播延时tp = (tpHL+tpLH)/2

* **功耗和能耗**
    * 功耗：
        * 动态功耗：开关瞬间，电容充分电引起的或电源到地的短暂电流通路引起的。频率越高，动态功耗越大
        * 静态功耗: 电源到地的静态电流引起的
    * 功耗延时积：
        $E_{PDP} = P_{av} \times t_p$ ($E_{PDP}$一般为常数 )
    * 功率：
        1. 瞬时功率 $ p(t) = v(t)i(t)$
        2. 峰值功率 $ p_{peak} = V_{supply}i_{peak}$
        3. 平均功率 $ P_{ave} = \frac{1}{T} \int^{t+T}_t p(t)dt = \frac{V_{supply}}{T} \int^{t+T}_t i_{supply}(t)dt $
