## 前言
* 指令系统是计算机系统结构的主要组成部分
* 指令系统是软件和硬件分界面的一个主要标志
* 指令系统是软件和硬件之间互相沟通的桥梁
* 指令系统与软件之间的语义差距越来越大

## 数据表示
* 数据表示与数据类型
    * 数据表示：计算机**硬件**能够**直接识别、可以被指令系统直接调用**的数据类型
        * 常见的数据类型由：定点数，逻辑数，浮点数，十进制数，字符，字符串，堆栈和向量等   
        * *确定哪些数据类型用数据表示实现，是软件和硬件的取舍问题*
    * **确定用数据表示的原则**
        * **缩短程序的运行时间**
        * **减小CPU与主存储器之间的通信量**
        * **这种数据表示的通用性和利用率**
* 浮点数的设计方法
    * 浮点数的表示方法: $ N = m·r^{e}_m $，其中 $ e = r^q_e $
        * 尾数的值：m -----包括码制(原、补码)和数制（小数、整数）
        * 阶码的值：e -----一般采用移码或补码，数制为整数
        * 尾数的基：r_m ---2进制，4进制，8进制，16进制，10进制等
        * 阶码的基：r_e ---通常为2
    * 浮点数的存储方式：
        | 1位 | 1位 | q位 | p位 |
        | --- | --- |---|---|
        |$m_f$|$e_f$|  e | m |
        * $m_f,e_f$为尾数和阶码的符号位
        * e为阶码的值(g)，m为尾数的值
    * 浮点数的表数范围：
        * 尾数为原码、小数，阶码为移码、整数时，
            * 规格化浮点数N的表数范围：$ r_m^{-1} · r_m^{-r^q_e} \leq |N| \leq (1-r^{-p}_m) · r^{r_e^q-1}_m $
            * 最值如下图所示
            
                <img src = "./浮点数范围.png" width = 80%>
        * 尾数用补码，纯小数时的表述范围：
            * 负数: $ -r_m^{r_e^q-1} \leq N \leq -(r_m^{-1} + r_m^{-p})·r_m^{-r_e^q} $
            * 正数: $ r_m^{-1} · r_m^{-r_m^{q}} \leq N \leq (1 - r_m^{-p})·r_m^{-r_e^q - 1} $
    * 设计参数的确定原则：
        1. 尾数：采用原码，小数表示   ------> 便于乘除运算
        2. 阶码：采用整数，移码表示
        3. 尾数的基数$r_m$选择2
        4. 阶码的基数$r_d$取2

* 原码，反码，补码，移码
    * 正数：原码 = 反码 = 补码；
    * 负数：
\begin{align*}
        原码 & = 1 + 真值\\
        反码 &= 1 + 真值的反\\
        补码 &= 1 + (真值的反 + 1 )
\end{align*}
    * 移码： 将补码的符号位取反
* 数值长度为n，原码、反码、补码的表数范围：
    * 原码和反码的表数范围：
        * 整数：$-(2^n-1) \leq N \leq (2^n-1)$
        * 小数：$-(1-2^{-n}) \leq N \leq (1 - 2^{-n}) $
    * 补码的表数范围：
        * 整数：$-2^n \leq N \leq (2^n-1)$
        * 小数：$-1 \leq N \leq (1 - 2^{-n}) $

* 自定义数据表示
    * 一般处理机的数据表示方法：数据存储单元只存放纯数据，由指令中的操作码来解释
    * 带标志符的数据表示法：
        <img src = "./标志符的数据表示.png" width = 80%>
        
        * 标志符由编译器或其他系统软件建立，对程序员透明
        * 主要优点：
            * 简化了指令系统
            * 由硬件自动实现一致性检查和数据类型的转换
            * 简化程序设计，缩小了人与机器的语义差距
            * 简化编译器，缩小了高级语言与机器语言的语义差距
            * 支持数据库系统，一个软件不加修改就可适用于多种数据类型
            * 方便软件调试
        * 主要缺点：
            * 数据和指令的长度可能不一致
            * 指令的执行速度降低，程序的设计时间、编译时间和调试时间缩短
            * 硬件复杂度增加

            * 功能位：操作数，指令，地址，控制字
            * 陷井位：由软件定义四种捕捉方式
            * 封写位：指定数据是只读还是可读可写
            * 类型位：二进制，十进制，定点数，浮点数，复数等等
            
    * 数据描述符表示法：
        * 与标志符的区别：标志符只作用一个数据，而数据描述符要作用与一组数据
